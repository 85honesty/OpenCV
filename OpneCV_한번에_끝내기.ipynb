```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from PIL import Image
import requests
from io import BytesIO
```

```python
url = 'https://cdn.pixabay.com/photo/2018/10/01/09/21/pets-3715733__480.jpg'

response = requests.get(url)
pic = Image.open(BytesIO(response.content))
```

OpenCV의 채널 순서
- OpenCV를 통해 영상(이미지)를 다룰 때의 채널 순서는 B G R
- matplotlib은 R G B 순서

```python
cv2_imshow(pic_arr)
```

cv2.cvtColor()
- image arrary, 변경할 색공간을 인자로 넣어줌
- 변경할 색공간은 여러 가지가 있음
    - cv2.COLOR_BGR2RGB
    - cv2.COLOR_RGB2GRAY
    - cv2.COLOR_GRAY2RGB
    - ...

```python
image = cv2.cvtColor(pic_arr, cv2.COLOR_RGB2BGR)
cv2_imshow(image)
```
